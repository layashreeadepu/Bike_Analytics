```python
import os
import requests
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# 1. Setup Browser
website_url = "https://s3.amazonaws.com/hubway-data/index.html"
driver = webdriver.Chrome() 
driver.get(website_url)

# 2. Create a folder for the data if it doesn't exist
folder_name = 'hubway_downloads'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# 3. Wait for the JavaScript to load the file list
wait = WebDriverWait(driver, 10)
try:
    wait.until(EC.presence_of_element_located((By.XPATH, '//a[contains(@href, ".zip")]')))
    zip_links = driver.find_elements(By.XPATH, '//a[contains(@href, ".zip")]')
    print(f"Total files found: {len(zip_links)}")

    # 4. Loop and Download
    for link in zip_links:
        url = link.get_attribute("href")
        file_name = link.text
        file_path = os.path.join(folder_name, file_name) # Saves inside the folder
        
        print(f"Downloading {file_name}...")
        
        # Using stream=True is better for large zip files to avoid memory issues
        response = requests.get(url, stream=True)
        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
                
    print("--- All downloads complete! ---")

finally:
    driver.quit()
```
